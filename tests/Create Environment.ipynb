```python
config = {
    'population_size': 50,
    'max_validity_attempts': 200,
    'max_individual_size': 100,
    'max_number_of_targets': 10,
    'max_value': 9999999,
    'precision': 6,
    'fitness_function': 'MEAN_SQUARED_DEVIATION'
}
```

```python
from symbolic_regression import Environment
```

```python
model = Environment(config)
```

```python
for num, individual in enumerate(model.population):
    print(num)
    line = ''
    for item in individual.print(
        validity_attempts=True,
        genotype=False,
        chromossome=True
    ).split(';'):
        item = item.strip()
        if item.endswith(']'):
            print(item)
            continue
        item += ' '*(36-len(item))
        line += item
        if len(line) >= 90:
            print(line)
            line = ''
    print(line, '\n')
```

```python
model.fit(
    [1,  2,  3,  10, 20, 30, 13, 14, 15],
    [10, 11, 12, 19, 29, 39, 22, 23, 24]
)
```
